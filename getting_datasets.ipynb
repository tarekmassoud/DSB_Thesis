## Importing Necessary Libraries

```python
import pandas as pd
pd.set_option('display.max_columns', None)

import warnings
warnings.simplefilter("ignore", UserWarning)
```

## Reading The Datasets

```python
df_2013 = pd.read_excel('data/ATP_2013.xlsx')
df_2014 = pd.read_excel('data/ATP_2014.xlsx')
df_2015 = pd.read_excel('data/ATP_2015.xlsx')
df_2016 = pd.read_excel('data/ATP_2016.xlsx')
df_2017 = pd.read_excel('data/ATP_2017.xlsx')
df_2018 = pd.read_excel('data/ATP_2018.xlsx')
df_2019 = pd.read_excel('data/ATP_2019.xlsx')
df_2020 = pd.read_excel('data/ATP_2020.xlsx')
df_2021 = pd.read_excel('data/ATP_2021.xlsx')
df_2022 = pd.read_excel('data/ATP_2022.xlsx')
df_2023 = pd.read_excel('data/ATP_2023.xlsx')
df_2024 = pd.read_excel('data/ATP_2024.xlsx')
```

## Concatenating The Datasets

```python
# List of all the dataframes
dfs = [
    df_2013, df_2014, df_2015, 
    df_2016, df_2017, df_2018, 
    df_2019, df_2020, df_2021, 
    df_2022, df_2023, df_2024
]

# Concatenate all dataframes along the rows 
df_final = pd.concat(dfs, ignore_index=True)

# Sort the DataFrame by the Date column
df_final['Date'] = pd.to_datetime(df_final['Date']) 
df_final = df_final.sort_values(by='Date', ascending=True)

# Reset index after sorting
df_final = df_final.reset_index(drop=True)
```
